```python
#Finding maximum and minimum values
t1=(10,20,30,40,50,)
print("Maximum:",max(t1))
print("Minimum:",min(t1))
```

```python
#Count of an element
t1=(3,2,1,3,2,1,2,3,2,3)
res=t1.count(3)
print(res)
```

```python
#Finding length of tuple
t1=(8,5,3,9,6,4,2,1,7)
res = len(t1)
print(res)
```

```python
#Accessing tuple elements (indexing)
t1=(10,20,30,40,50)
print("first elelment:",t1[0])
print("second element:",t1[-1])
```

```python
#Finding index of an element
t1=(10,10,30,40)
print(t1.index(30))
```

```python
#Accessing element 
t1=(10,20,30,40)
print(t1[-1])
```

```python
#Slicing a tuple
t1=(10,20,30,40,50)
print(t1[1:4])
```

```python
#Membership (in and not in)
t1=(10,20,30,40)
print(20 in t1)
print(50 not in t1)
```
