```python
#print all keys
student={
    "name":"Sakshi",
    "age":20,
    "course":"AI"
}
print(student.keys())
```

```python
#print all values
student={
    "name":"Sakshi",
    "age":20,
    "course":"AI"
}
print(student.values())
```

```python
#print keys using loop
student={
    "name":"Sakshi",
    "age":20,
    "course":"AI"
}

for key in student:
    print(key)
```

```python
#print keys and values together
student={
    "name":"Sakshi",
    "age":20,
    "course":"AI"
}

for key,value in student.items():
    print(key,":",value)   
```

```python
#Access dictionary value
student={
    "name":"sakshi",
    "age":20
}
print(student["name"])
```

```python
#Add a new key-value pair
student={
    "name":"Sakshi",
    "age":20
}

student["city"]="pune"
print(student)
```

```python
#update a value
student={
    "name":"Sakshi",
    "age":20
}

student["age"]=21
print(student)
```

```python
#Delete a key
student={
    "name":"Sakshi",
    "age":20
}

del student["age"]
print(student)    
```

```python
#Count number of items
student={
    "name":"Sakshi",
    "age":20,
    "city":"Pune"
}

print(len(student))    
```

```python
#Copy a dictionary
student={"name":"sakshi","age":20}
new_student=student.copy()
print(new_student)
```

```python
#Clear a dictionary
student={"name":"sakshi","age":20}
student.clear()
print(student)
```

```python
#Merge two dictionaries
d1={"a":10,"b":20}
d2={"c":30,"d":40}
d1.update(d2)
print(d1)
```

```python
#Sum of dictionary values
marks={"Math":80,"Science":75,"English":90}
print(sum(marks.values()))
```

```python
#Find maximum value
marks={"Math":70,"Science":75,"English":80}
print(max(marks.values()))
```

```python
#Create dictionary from two lists
keys=["name","age","city"]
values=["sakshi",20,"pune"]
student=dict(zip(keys,values))
print(student)
```
